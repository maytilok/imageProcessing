```python
import pandas as pd
import numpy as np

# Specify the path to your Excel file in Google Drive
file_path = '/content/drive/MyDrive/Food Image Analysis/Analysis/Image/Kfc/Kfc_tweet_image_19_24_6.xlsx'

# Read the Excel file into a DataFrame
twitter_df = pd.read_excel(file_path)
```

```python
# Create a run number column
twitter_df['tweet_id'] = range(1, len(twitter_df) + 1)

# Set 'run_number' as the index
twitter_df = twitter_df.set_index('tweet_id')
```

```python
twitter_df_clean=twitter_df[['Relationship', 'Tweet', 'Assertive', 'Expressive', 'Retweet Count', 'Favorite Count',
                             'Reply Count', 'Quote Count', 'Impression Count', 'Hashtags in Tweet', 'Media in Tweet',
                             'Media Type', 'Twitter Page for Tweet', 'Tweet Date (UTC)', 'Date', 'Time', 'Tweet Image File']]
```

```python
twitter_df_clean = twitter_df_clean.rename(columns={'Assertive': 'Assertive_1', 'Expressive': 'Expressive_1',
                                                    'Retweet Count': 'Retweet_Count', 'Favorite Count': 'Favorite_Count',
                                                    'Reply Count': 'Reply_Count', 'Quote Count': 'Quote_Count',
                                                    'Impression Count': 'Impression_Count', 'Hashtags in Tweet':'Hashtags_in_Tweet',
                                                    'Media in Tweet': 'Media_in_Tweet', 'Media Type': 'Media_Type',
                                                    'Twitter Page for Tweet': 'Twitter_Page_for_Tweet', 'Tweet Date (UTC)': 'Tweet_Datetime',
                                                    'Tweet Image File': 'Tweet_Image_File'})
```

```python
def split_urls_in_dataframe(df, column_name, separator=' '):
    """
    Splits a column in a DataFrame containing multiple URLs into multiple rows.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column containing URLs.
        separator (str, optional): The separator between URLs. Defaults to ', '.

    Returns:
        pd.DataFrame: A new DataFrame with URLs split into separate rows.
    """
    df_new = df.copy()
    df_new[column_name] = df_new[column_name].str.split(separator)
    df_new = df_new.explode(column_name)
    return df_new
```

```python
Twitter_df_url_split = split_urls_in_dataframe(twitter_df_clean, 'Media_in_Tweet')
```

```python
Twitter_df_url_split
```

```python
image_check = ['.jpg', '.png']
Twitter_df_url_split_clean = Twitter_df_url_split[Twitter_df_url_split['Media_in_Tweet'].str.endswith(tuple(image_check))]
```

```python
subset_columns = ['Relationship','Tweet','Assertive_1','Expressive_1', 'Retweet_Count',
                  'Favorite_Count','Reply_Count','Quote_Count','Impression_Count','Hashtags_in_Tweet',
                  'Media_in_Tweet', 'Media_Type', 'Twitter_Page_for_Tweet', 'Tweet_Datetime', 'Date',
                  'Time','Tweet_Image_File' ]
df_unique = Twitter_df_url_split_clean.drop_duplicates(subset=subset_columns)
```
