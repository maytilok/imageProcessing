###Consummatory

```python
# Function to determine comsummatory based on condition
def condition_value(row):
    if row['face'] == 1 & row['food_objects']==1:
        return 1
    else:
        return 0
```

```python
subset_columns = ['brightness', 'face', 'Relationship','Tweet','Assertive_1','Expressive_1', 'Retweet_Count',
                  'Favorite_Count','Reply_Count','Quote_Count','Impression_Count','Hashtags_in_Tweet',
                  'Media_in_Tweet', 'Media_Type', 'Twitter_Page_for_Tweet', 'Tweet_Datetime', 'Date',
                  'Time','Tweet_Image_File', 'Media_Tweet_url', 'color_variation', 'object_name',
                  'food_objects','bundle', 'consummatory', 'resolution']
df_unique = TweetFace_bright_color_label_reso_df.drop_duplicates(subset=subset_columns)
```

```python
df_unique
```

```python
df_unique['new_index'] = pd.RangeIndex(start=1, stop=len(df_unique) + 1)
df_unique = df_unique.set_index('new_index')
```

```python
df_unique
```

```python
#Drop the index column
df_drop_index = TweetFace_bright_color_label_df.reset_index(drop=True, inplace=True)
```

```python
#Group by 'ID' and merge measure value
df_merged = df_unique.groupby(['tweet_id']).agg({'face': 'sum', 'consummatory':'sum', 'brightness': 'mean',
                                                 'resolution': 'mean','color_variation':'mean','food_objects':'sum',
                                                 'bundle':'sum', 'object_name': ','.join})
```

```python
df_merged
```

```python
#if sum>1 meaning 1
df_merged['Face'] = df_merged.apply(lambda row: 1 if row['face'] >= 1 else 0, axis=1)
df_merged['Consummatory'] = df_merged.apply(lambda row: 1 if row['consummatory'] >= 1 else 0, axis=1)
df_merged['Food_objects'] = df_merged.apply(lambda row: 1 if row['food_objects'] >= 1 else 0, axis=1)
df_merged['Bundle'] = df_merged.apply(lambda row: 1 if row['bundle'] >= 1 else 0, axis=1)
```

```python
df_merged.rename(columns={'brightness':'Brightness', 'resolution':'Resolution',  'color_variation':'Color_variation', 'object_name':'Object_name'}, inplace=True)
```

```python
df_merged_clean = df_merged.drop(['face','consummatory', 'food_objects', 'bundle'], axis=1)
```

```python
df_merged_clean.dtypes
```

```python
Tweet_measure_df = df_merged_clean.join(df_unique, how = 'inner')
```
