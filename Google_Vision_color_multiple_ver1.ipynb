```python
# Import the Vision client from Google Cloud
from google.cloud import vision
from google.colab import auth
auth.authenticate_user(project_id='label-mcdonald-test')
```

```python
import pandas as pd
import cv2
import os
```

```python
def detect_properties(image_path):
  color_variation=[]

  for path in image_path:
    """Detects image properties in the file."""
    client = vision.ImageAnnotatorClient()

    with open(path, "rb") as image_file:
        content = image_file.read()

    image = vision.Image(content=content)

    response = client.image_properties(image=image)
    props = response.image_properties_annotation
    #print("Properties:")

    new_list=list()
    for color in props.dominant_colors.colors:
        #print(f"fraction: {color.pixel_fraction}")
        #print(f"score: {color.score}")
        new_list.append(f"{color.score}")
    #print(new_list)
    score = [float(i) for i in new_list]
    total_sum = sum(score)
    percentages = [(number / total_sum) * 100 for number in score]
    top_three = sorted(percentages, reverse=True)[:3]
    colorful = sum(top_three)
    color_variation.append(colorful)
    #print(percentages)
    #print(top_three)
    #print(color_variation)

  return color_variation
```

```python
#if __name__ == "__main__":
    # List of image paths (you can modify this to your own image paths)
image_folder = "/content/drive/MyDrive/VGG16/McDonalds_Image_NodeXL_19_24"
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

colorful_results = detect_properties(image_files)
# Initialize an empty list to store the data
data_list = []

# Print results
for img, result in zip(image_files, colorful_results):
    data = {'image': img, 'color_variation': result}
    data_list.append(data)  # Append the dictionary to the list
    #print(f"{img}: {'Face detected' if result == 1 else 'No face detected'}")
```

```python
# Convert the list of dictionaries to a DataFrame
color_df = pd.DataFrame(data_list)
color_df
```
