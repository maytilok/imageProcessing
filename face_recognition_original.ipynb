```python
import face_recognition
import cv2
import os

def detect_faces(image_paths):
    results = []

    for image_path in image_paths:
        # Load the image file
        image = face_recognition.load_image_file(image_path)

        # Find all face locations in the image
        face_locations = face_recognition.face_locations(image)

        # Check if any faces were found
        if len(face_locations) > 0:
            results.append(1)  # Face detected
        else:
            results.append(0)  # No face detected

    return results
```

```python
import pandas as pd

#if __name__ == "__main__":
    # List of image paths (you can modify this to your own image paths)
image_folder = "/content/drive/MyDrive/VGG16/McDonalds_Image_NodeXL_19_24"
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

face_detection_results = detect_faces(image_files)
# Initialize an empty list to store the data
data_list = []

# Print results
for img, result in zip(image_files, face_detection_results):
    data = {'image': img, 'face': result}
    data_list.append(data)  # Append the dictionary to the list
    #print(f"{img}: {'Face detected' if result == 1 else 'No face detected'}")
```

```python
# Convert the list of dictionaries to a DataFrame
face_df = pd.DataFrame(data_list)
face_df
```

```python
#count image
face_detected_sum = face_df['face'].sum()
no_face_detected_sum = len(face_df) - face_df['face'].sum()
```

```python
#display result
summary_df = pd.DataFrame({'Face_detected':[face_detected_sum], 'No_face_detected':[no_face_detected_sum]})
summary_df
```
