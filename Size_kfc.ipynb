```python
def get_image_resolution(image_path):
  results = []

  for path in image_files:
    img = Image.open(path)
    # get image size
    width, height = img.size
    # calculate resolution
    resolution = width*height
    results.append(width)
  return results
```

```python
import cv2
import os
import pandas as pd

image_folder = "/content/drive/MyDrive/Food Image Analysis/Analysis/Image/Kfc/Kfc_Image"
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

resolution_results = get_image_resolution(image_files)

# Initialize an empty list to store the data
resolution_list = []

# Print results
for img, result in zip(image_files, resolution_results):
    data = {'image': img, 'resolution': result}
    resolution_list.append(data)  # Append the dictionary to the list
```

```python
import pandas as pd

resolution_df = pd.DataFrame(resolution_list)
resolution_df
```
