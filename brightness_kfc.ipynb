```python
def cal_brightness(image_paths):

  results = []

  for path in image_files:
    img = Image.open(path)
    # Convert to grayscale
    gray_img = img.convert('L')
    # Convert to numpy array
    gray_array = np.array(gray_img)
    # Calculate mean brightness
    brightness = np.mean(gray_array)
    results.append(brightness)
  return results
```

```python
import cv2
import os
import pandas as pd

image_folder = "/content/drive/MyDrive/Food Image Analysis/Analysis/Image/Kfc/Kfc_Image"
image_files = [os.path.join(image_folder, f) for f in os.listdir(image_folder) if f.endswith(('.jpg', '.png'))]

brightness_results = cal_brightness(image_files)

# Initialize an empty list to store the data
brightness_list = []

# Print results
for img, result in zip(image_files, brightness_results):
    data = {'image': img, 'brightness': result}
    brightness_list.append(data)  # Append the dictionary to the list
```

```python
import pandas as pd

brightness_df = pd.DataFrame(brightness_list)
brightness_df
```
